3. Правда ли, что люди с высоким уровнем стресса имеют более высокое кровяное давление, чем те, кто не подвержен стрессу

````python
stress_column = 'Stress Levels'
pressure_column = 'Blood Pressure (s/d)'

pressure_parts = df[pressure_column].str.split('/', expand=True)
df['Systolic'] = pd.to_numeric(pressure_parts[0], errors='coerce')
df['Diastolic'] = pd.to_numeric(pressure_parts[1], errors='coerce')

bins = [0,3,6,10]
labels = ['Низкий', 'Средний', 'Высокий']
df['Stress Group'] = pd.cut(df[stress_column], bins=bins, labels=labels, include_lowest=True)

result = df.groupby('Stress Group', observed=True).agg(
    avg_systolic=('Systolic', 'mean'),
    avg_diastolic=('Diastolic', 'mean'),
    median_systolic=('Systolic', 'median'),
    median_diastolic=('Diastolic', 'median'),
    person_count=('Stress Levels', 'count')
)
print(result)

low_stress_avg_sys = result.loc['Низкий', 'avg_systolic']
high_stress_avg_sys = result.loc['Высокий', 'avg_systolic']

low_stress_avg_dys = result.loc['Низкий', 'avg_diastolic']
high_stress_avg_dys = result.loc['Высокий', 'avg_diastolic']

if high_stress_avg_sys > low_stress_avg_sys and high_stress_avg_dys > low_stress_avg_dys:
    print("Да, у людей с выскоким уровнем стресса более высокое кровяное давление")
else:
    print("Нет, у людей с высоким уровнем стресса не больше кровяное давление чем у людей с низким уровнем стресса")
````

4. Правда ли, что уровень холестерина у мужчин и женщин отличается?

````python
cholesterol_col = 'Cholesterol Level (mg/dL)'
men_cholesterol = df[df['Gender'] == 'Male'][cholesterol_col]
women_cholesterol = df[df['Gender'] == 'Female'][cholesterol_col]

avg_men = men_cholesterol.mean()
avg_women = women_cholesterol.mean()

print(df.groupby('Gender')[cholesterol_col].describe())

if (avg_men > avg_women):
    print('Да, в среднем у мужчин больше холестерина')
elif (avg_women == avg_men):
    print('Нет, средний уровень холестерина одинаковый у обоих полов')
else:
    print('Да, в среднем у женщин больше холестерина')

````

5. Постройте круговую диаграмму отображающую как ментальное здоровье связано с паттернами сна

````python
mental_health_col = 'Mental Health Status'
sleep_col = 'Sleep Patterns'

if mental_health_col not in df.columns or sleep_col not in df.columns:
    print(f"❌ Ошибка: Одна из колонок ('{mental_health_col}' или '{sleep_col}') не найдена в файле.")
    exit()

health_order = ['Poor', 'Fair', 'Good', 'Excellent']
df[mental_health_col] = pd.Categorical(df[mental_health_col], categories=health_order, ordered=True)

sns.set_theme(style="whitegrid")

fig, axes = plt.subplots(1, len(health_order), figsize=(24, 7))
fig.suptitle('Распределение паттернов сна в зависимости от ментального здоровья', fontsize=20, fontweight='bold')

colors = sns.color_palette('viridis', n_colors=len(df[sleep_col].unique()))

for i, status in enumerate(health_order):
    subset = df[df[mental_health_col] == status]
    
    sleep_counts = subset[sleep_col].value_counts()
    
    axes[i].pie(sleep_counts, labels=sleep_counts.index, autopct='%1.1f%%', 
                startangle=90, colors=colors, wedgeprops={'edgecolor': 'white'})
    axes[i].set_title(f'Здоровье: {status}', fontsize=14)
    axes[i].set_ylabel('')

plt.tight_layout(rect=[0, 0, 1, 0.95])
plt.show()
````

6. Имеет ли связь зрение и способность слышать звуки на высоких частотах? Ответьте на вопрос - какая зависимость между ними и предположите почему зависимость именно такая?

````python
vision_col = 'Vision Sharpness'
hearing_col = 'Hearing Ability (dB)'

if vision_col not in df.columns or hearing_col not in df.columns:
    print(f"❌ Ошибка: Одна из колонок ('{vision_col}' или '{hearing_col}') не найдена в файле.")
    exit()


sns.set_theme(style="whitegrid")
plt.figure(figsize=(12, 7))

sns.regplot(data=df, x=vision_col, y=hearing_col,
            scatter_kws={'alpha': 0.5, 'color': 'green'},
            line_kws={'color': 'orange', 'linestyle': '--'})

plt.title('Связь между остротой зрения и слухом', fontsize=16)
plt.xlabel('Острота зрения', fontsize=12)
plt.ylabel('Способность слышать (дБ)', fontsize=12)
plt.show()


correlation = df[vision_col].corr(df[hearing_col])

print("\n--- Числовой анализ ---")
print(f"Коэффициент корреляции между остротой зрения и слухом: {correlation:.4f}")



print("\n--- Вывод и предположение ---")
if correlation < -0.7:
    print(f"Обнаружена сильная отрицательная связь ({correlation:.2f}).")
    print("   Это означает, что чем выше острота зрения, тем хуже человек слышит высокочастотные звуки (и наоборот).")
elif correlation < -0.4:
    print(f"Обнаружена заметная отрицательная связь ({correlation:.2f}).")
    print("Существует четкая тенденция: люди с лучшим зрением, как правило, хуже слышат высокие частоты.")
elif correlation > 0.4:
    print(f"Обнаружена заметная положительная связь ({correlation:.2f}).")
    print("Это означает, что люди с хорошим зрением, как правило, хорошо слышат и высокие частоты.")
else:
    print(f"Связь слабая или отсутствует ({correlation:.2f}).")
    print("На основе этих данных нельзя утверждать о наличии сильной линейной зависимости между зрением и слухом.")

print("Связь скорее всего в возрасте, чем моложе организм - тем четче он слышит и видит")
````
